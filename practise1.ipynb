```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Kinase_final_data.csv')
df
```

```python
df.dtypes
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
sb.heatmap(df.isnull())
```

```python
x=df.drop(['molecule_chembl_id','canonical_smiles','pIC50'],axis=1)
y=df['pIC50']
```

```python
x
```

```python
y
```

```python
df_num=df.select_dtypes(['float64','int64'])
```

```python
df_num
```

```python
df_obj=df.select_dtypes(object)
```

```python
df_obj
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df_obj.columns
for i in df_obj:
    df_obj[i]=le.fit_transform(df_obj[i])
```

```python
df_obj
```

```python
df_sum=pd.concat([df_obj,df_num],axis=1)
```

```python
df_sum
```

```python
x=df_sum.drop(['molecule_chembl_id','canonical_smiles','pIC50'],axis=1)
y=df_sum['pIC50']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(units=7,activation='relu',input_shape=(x.shape[1],)),
    #output layer
    tf.keras.layers.Dense(units=1,activation='linear')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',loss='mse')
```

```python
#create procedure to decide number of iterations or epochs
#first early stopping class
from tensorflow.keras.callbacks import EarlyStopping 
#create a object of early stopping class
cb=EarlyStopping(
    monitor="val_loss",# val loss validation loss means testing error
    min_delta=0.00001,  #delta min learning rate
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None, #nueral network decided automatic
    restore_best_weights=False #default true but false it will assign first weights

)
```

```python
trained_model=model.fit(x_train,y_train,epochs=5000,callbacks=cb,validation_data=(x_test,y_test))
```

```python
#visulaised loss and epoch
plt.plot(trained_model.history['loss'],label='Training loss')
plt.xlabel("No. of Epochs")
plt.ylabel('Mean square error')
plt.legend()
plt.show()
```

```python
plt.plot(trained_model.history['loss'],label='Training loss')
plt.plot(trained_model.history['val_loss'],label='Testing loss')
plt.xlabel("No. of Epochs")
plt.ylabel('Mean square error')
plt.legend()
plt.show()
```

```python
model.get_weights() # weight #bais 2 layer weight bais
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
print("training loss",model.evaluate(x_train,y_train))
print("testing loss",model.evaluate(x_test,y_test))
```
